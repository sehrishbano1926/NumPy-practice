**First Dimension (Rows): 1:3**

1 is the start index (inclusive), so it starts from the second row (index 1).


3 is the stop index (exclusive), so it stops before the fourth row (index 3).


This selects the second and third rows.


**Second Dimension (Columns): 1:2**

1 is the start index (inclusive), so it starts from the second column (index 1).

2 is the stop index (exclusive), so it stops before the third column (index 2).
This selects only the second column.

So, array_t[1:3, 1:2] extracts the elements from rows 2 and 3, but only the second

# **image processing**



```python
import numpy as np
import imageio
#image of room
room_image=imageio.imread('/content/room.avif')
#displaying the shape of the image
room_image.shape
```

```python
import matplotlib.pyplot as plt
#displaying the image
plt.imshow(room_image)
#hide the aixs
plt.axis("off")
plt.show()
```

```python
#convert the image to grayscale
grayscale_image=room_image[:, :, 0]* 0.1334 + room_image[:, :,1]* 0.2384 + room_image[:, :, 2]*0.3451
# display the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.axis("off")
plt.show()
```

```python
#computing the negative of the grayscale image
negative_image=255- grayscale_image
#display the negative image
plt.imshow(negative_image, cmap='gray')
plt.axis('off')
plt.show
```

```python
#applying thresholding to the grayscale image
threshold_value=128
binary_image=(grayscale_image > threshold_value)* 255

#display the binary image
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
plt.show
```

```python
# computing the histogram of the grayscale image
histogram, bin_edges=np.histogram(grayscale_image, bins=256, range=(0, 256))
#plotting the histogram
plt.figure(figsize=(8, 4))
plt.bar(bin_edges[:-1], histogram, width=1, color='gray')
plt.title('Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('number of Pixels')
plt.show()
```

```python
#Normalizing the grayscale
normalized_image=(grayscale_image - np.min(grayscale_image)) / (np.max(grayscale_image))

#displyaing the normalized image
plt.imshow(normalized_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#flipping the grayscale image

flipped_horizontal=np.fliplr(grayscale_image)
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

#display the orignal and flipped images
ax[0].imshow(grayscale_image, cmap='gray')
ax[0].axis('off')
ax[0].set_title('original image')

ax[1].imshow(flipped_horizontal, cmap='gray')
ax[1].axis('off')
ax[1].set_title('horizontally fillped')

plt.tight_layout()
plt.show()


```

```python
from scipy.ndimage import rotate
#rotate the grayscale_image by any degree
rotated_image=rotate(grayscale_image, angle=45, mode='reflect')

#display

plt.imshow(rotated_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# defining the cordinates for cropping(y1, y2, x1, x2)

y1, y2, x1, x2=100, 400, 100 ,400

#cropping the image
cropped_image=grayscale_image[y1:y2, x1:x2]

#displaying the cropped_image
plt.imshow(cropped_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
from scipy.ndimage import zoom

height=200
width=200

grayscale_image.shape[0]
grayscale_image.shape[1]



#resize_image

resize_image=zoom(grayscale_image, (height, width))

#diaplay

plt.imread(resize_image, cmap='gray')

plt.axis('off')
plt.show()
```
